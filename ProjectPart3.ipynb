## 1 Graphs/Trees

A structure with at least 5 vertices and 5 edges, based on something from my life.  In this case, I am choosing to construct a simple family tree.

```
import graphviz

Millers = graphviz.Digraph('MillerFamilyTree')

Millers.node('Great-Grandpa Miller')
Millers.edge('Great-Grandpa Miller','Curt')
Millers.edge('Curt','Darlene',label='Married')
Millers.edge('Curt', 'Dennis','Son')
Millers.edge('Darlene', 'Dennis','Son')
Millers.edge('Curt', 'Curtis','Son')
Millers.edge('Darlene', 'Curtis','Son')
Millers.edge('Dennis','Samuel','Son')
Millers.edge('Curt', 'Samuel','Grandson')
Millers.edge('Darlene', 'Samuel','Grandson')
Millers.edge('Curtis', 'Samuel','Nephew')
Millers.edge('Sheri','Dennis','Married')
Millers.edge('Sheri','Samuel','Son')
Millers
```

A bit confusing with all the lines, but basically:

- Great-grandpa Miller and his wife had my grandfather and his siblings
-My grandfather married my grandmother
-Grandparents had two sons; my dad & his brother my uncle
-My father and mother had me

While both my mother and grandmother were not originally part of the Miller family tree, I added them to both fill out the graph to a more complete state, as well so I could add a couple lables others than just son or grandson.
